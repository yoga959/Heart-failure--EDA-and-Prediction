```python
x_train=data.iloc[:,:12]
x_test=data.iloc[:,:12]

y_train=data['DEATH_EVENT']
y_test=data['DEATH_EVENT']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_train,y_train,test_size=0.20,random_state=43)
```

##### Finding imp feature based on random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
R=RandomForestClassifier()
R.fit(x_train,y_train)
prediction=R.predict(x_test)
print("The accuracy by using original dataset:",metrics.accuracy_score(prediction,y_test))
print("Accuracy by using original dataset:",metrics.precision_score(prediction,y_test))
print("Accuracy by using original dataset:",metrics.recall_score(prediction,y_test))
```

```python
feature_importances = R.feature_importances_
print(feature_importances)
```

```python
indices = np.argsort(feature_importances)[::-1]
num_features = 10
top_indices = indices[:num_features]
top_importances = feature_importances[top_indices]



print("Top 10 feature rankings based on random forest:")
for f in range(num_features): 
    print(f"{f+1}. {x_train.columns[indices[f]]}: {feature_importances[indices[f]]}")

plt.barh(range(num_features), top_importances, align="center")
plt.yticks(range(num_features), x_train.columns[top_indices])
plt.xlabel("Feature Importance based on random forest")
plt.ylabel("Feature")
plt.show()
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(x_new_train,y_new_train)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(x_new_train.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(10,'Score'))  
```

```python
import plotly.express as px
fig = px.bar(
    featureScores.nlargest(10, 'Score'),  
    x='Specs', 
    y='Score', 
    text='Score', 
    title='Top 10 Features Based on Chi-Squared Scores',
    labels={'Specs': 'Feature', 'Score': 'Chi-Squared Score'},
)
fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')
fig.update_layout(
    xaxis_tickangle=45,  
    height=500,
    width=800,
)
fig.show()
```
